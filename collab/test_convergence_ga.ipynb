```python
import json
import matplotlib.pyplot as plt

# 1. Buka brankas JSON khusus GA (pakai ../ karena posisi file ini di dalam folder collab)
path_ga = '../output_json/rute_ga.json'

with open(path_ga, 'r') as f:
    data_ga = json.load(f)

# 2. Siapkan kanvas grafik
plt.figure(figsize=(12, 6))

# 3. Looping untuk menggambar garis tiap klaster
warna_klaster = ['red', 'blue', 'green', 'orange', 'purple']
i = 0

for klaster, isi in data_ga.items():
    riwayat = isi.get("riwayat_konvergensi", [])
    
    # Cek apakah riwayatnya ada isinya atau kosong
    if len(riwayat) > 0:
        plt.plot(riwayat, label=f'Klaster {klaster.capitalize()}', color=warna_klaster[i], linewidth=2)
    else:
        print(f"⚠️ Peringatan: Riwayat konvergensi klaster {klaster} masih kosong! Cek lagi kodingan algoritma GA-nya.")
    
    i += 1

# 4. Percantik Grafik
plt.title('Grafik Konvergensi Genetic Algorithm (GA) - 5 Wilayah Surabaya', fontsize=14, fontweight='bold')
plt.xlabel('Generasi / Iterasi', fontsize=12)
plt.ylabel('Total Jarak (KM)', fontsize=12)
plt.legend(loc='upper right')
plt.grid(True, linestyle='--', alpha=0.7)

# 5. Tampilkan!
plt.tight_layout()
plt.show()
```
